```python
class Sorting():
    
    def __init__(self):
        pass
    
    def insertion_sort(self, to_sort):
        # Implementation von Insertion Sort
        # https://en.m.wikipedia.org/wiki/Insertion_sort
        
        # Stellt euch vor ihr hättet beim Kartenspielen 5 Spielkarten zugeteilt bekommen, die ihr nun sortieren wollt.
        # Ihr führt Insertion Sort aus, sofern ihr dabei sukzessive eine Karte aufnehmt, sie mit den bereits in eurer
        # Hand sortierten Karten vergleicht (von rechts nach links) und die Karte dann dort einsortiert, wo sie hinge-
        # hört - also an die Stelle, ander der alle Karte rechts von ihr großer und alle links von ihr kleiner gleich
        # sind.
        
        test_input_arr = list(to_sort)
        
        # Beginn mit dem ersten Kartepaar, da eine Karte trivialerweise sortiert ist.
        for i in range(1,len(to_sort)):
            # Aufnehmen der Karte, deren richtigen Einsortierort ihr bestimmen wollt und setzt sie ans rechte 
            # Ende eurer Hand
            put_in_place = test_input_arr[i]
            # Ihr beginnt den Vergleich beim Index -1 der Karte, die ihr gerade aufgenommen habt: siehe die 
            # nächsten zwei Zeilen
            j = i
            # die Schleife läuft solange wie noch Karten für den Vergleich (j > 0) übrig sind und solange die 
            # Karte j-1 größer der gleich der einzufügenden Karte (A[i] bzw. A[j]) ist.
            while j > 0 and test_input_arr[j-1] >= put_in_place:
                # Sind die Bedingungen der Schleife erfüllt, so vertauscht ihr jeweils die j-te mit der j-1-ten
                # Karte
                test_input_arr[j] = test_input_arr[j-1]
                # und geht für den nächsten Vergleich einen Karte weiter nach links in eurer Hand
                j = j - 1
            # sofern etweder die Karte, die ihr gerade in der Hand habt die neue kleinste Karte ist (bei j = 0),
            # fügt ihr diese am Anfang, also ganz links in eure Hand ein, ansonsten, fügt ihr sie genau rechts
            # von der Karte in eure Hand ein, die kleiner ist, als die Karte, die ihr nun einsortiert (genau dann,
            # wenn test_input_arr[j-1] < put_in_place gilt)
            test_input_arr[j] = put_in_place
        return ''.join(test_input_arr)
    
    
        def quicksort():
            #
            # https://de.wikipedia.org/wiki/Quicksort
            pass
    
    
        def mergesort():
            #
            # https://de.wikipedia.org/wiki/Mergesort
            
            pass
    
    
        def heapsort():
            #
            # https://de.wikipedia.org/wiki/Heapsort
            pass
                
```
